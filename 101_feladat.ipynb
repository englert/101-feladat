**Tipp** Mindig használd a `return` utasítást a függvényeidben. Az assert állítások automatikusan meghívják a függvényeidet.

```python
# Futtasd ezt a cellát, hogy generáljunk néhány számot a későbbi függvényeinkhez.
import random

pozitiv_paros_szam = random.randrange(2, 101, 2)
negativ_paros_szam = random.randrange(-100, -1, 2)

pozitiv_paratlan_szam = random.randrange(1, 100, 2)
negativ_paratlan_szam = random.randrange(-101, 0, 2)
print("Most már van néhány véletlenszámunk a későbbi gyakorlatokhoz.")
print("A véletlen pozitív páros szám:", pozitiv_paros_szam)
print("A véletlen pozitív páratlan szám:", pozitiv_paratlan_szam)
print("A véletlen negatív páros szám:", negativ_paros_szam)
print("A véletlen negatív páratlan szám:", negativ_paratlan_szam)
```

```python
# Példa függvény definíció:
# Írj egy koszont függvényt, ami hozzáadja a "Hello, " sztringet az elejéhez és "!"-ot a végéhez bármely megadott bemenetnek.
def koszont(nev):
    return "Hello, " + nev + "!"

assert koszont("Jane") == "Hello, Jane!", "Kérlek ellenőrizd a bemeneteket és adattípusokat"
assert koszont("Pat") == "Hello, Pat!", "Kérlek ellenőrizd a bemeneteket és adattípusokat"
assert koszont("Astrud") == "Hello, Astrud!", "Kérlek ellenőrizd a bemeneteket és adattípusokat"
print("A példa függvény definíció megfelelően futott")
```

```python
# Egy másik példa függvény definíció:
# Ez a plusz_ketto függvény egy változót vesz fel és hozzáad 2-t.
def plusz_ketto(szam):
    return szam + 2

assert plusz_ketto(3) == 5
assert plusz_ketto(0) == 2
assert plusz_ketto(-2) == 0
print("A plusz_ketto állítások megfelelően futottak... A második függvény definíció példa megfelelően futott.")
```
